## Chapter 2 - Theory

Here we will look over some of the theory used in this assignment, as well as some function and formulas\
**Functions and Formulas**

Linear demand function - $P = a-bQ$\
This is used the inverse demand function, and it is used to show price related to quantity. Here Q is qunatity of demand, a is maximum price consumers are wulling to pay, and b is ther to indicate the rate at which demand falls as price increases.

Marginal Revenue for linear demand - $MR = a-2bQ$\
Marginal revenue declines twice as fast as price increases due to the linear demand curve's nature.

Profit Maximization - $MR = MC$\
Occurs when $MR=MC$, where $MC$ is the marginal cost of producing one more unit. This condition finds the quantity at which the firm maximizes its profit.

Consumer Surplus - $CS = \frac{1}{2}*Q(a/b-P) P$\
It measures the difference between what consumers are willing to pay and whay they actually pay.

Producer Surplus - $PS = (P-c)Q$\
Difference betweeen what producers receive for selling a good versus their cost of producing it.

Total Welfare - $Total\ Welfare = CS + PS$\
Total welfare is the total of $CS$ and $PS$, and it is the overall economic benefit to society.

**Monopoly Structure** \
A monopoly is a market with a single seller, selling a unique product in the market, regardless if it is by invention, law, or other. In a monopoly there are zero competition and no subtsituteion. Monopolies have a few key characteristics:
- Single Producer:\
  There is only one producer as I mention in that market that controles the supply.
- No Substitution:\
  The product that is offered has no close substitutes, creating a sole source for the product 
- Price Maker:\
  Because the control over the price they have, they can set the price of its product, changing the price by altering the quantity supplied.
- Barriers to entry:\
  For a monopoly to stay a monopoly it needs high barriers to enter, this can be legal, tech, or natural control over resources. 

In determining the profit-maximizing output and price a monopolist relies on the intersection of marginal revenue ($MR$) and marginal cost ($MC$). The linear demand function: $P=a-bQ$, where $P$ is the price, $Q$ is the quantity demanded, $a$ is the maximum price consumers are willing to pay, and $b$ indicates the rate at which demand galles as the price increses. The marginal revenue for a monopolist is given by $MR = a-2bQ$, so a demand curve that is twice as steep. This is becaise to sell an additional unit a monopolist must lower the price not just for that unit but for all previous units sold as well, leading to marginal revenue curve beeing 2x steeper.

MR = MC, the monopolist decisison on the quantity to produce. The quantity where the marginal revenue from selling an additional unit equals the marginal cost of producing it. The price is then determined from the demand curve at this quantity.

Monopoly pricing extend to consumer surplus, producer surplus, and total welfare. Consumer surplus is the difference between what consumers are willing to pay and what they actually pay. Producer surplus is the difference between what producers receive and their cost of production. In a competitive market, the equilibrium maximizes total welfare—the sum of consumer and producer surplus. In a monopoly however, the higher price and lower output compared to competitive markets lead to a loss in consumer surplus, an increase in producer surplus, but a net loss in total welfare. This loss is called deadweight loss wich represents the inefficiency introduced by monopoly.  It occurs because there are potential transactions that would have been mutually beneficial like buyers valuing the product more than it cost to produce, that dooes not happen under a monopoly. The deadweight loss is illustrated as the area between demand and supply or marginal cost curve here. This then leads to a redistribution of welfare from consumers to the monopolist, with a portion of consumer surplus converted into producer surplus and another portion lost entirely as deadweight loss. 

**Third Degree Price Discrimination** \
First, there are multiple different price discrimination, but we will look at third degree. Third degree price discrimination occurs when a seller charges different prices to different groups of consumers for the same good or service based on each groups willingness to pay. This is used in multiple industries to maximize profits by capturing more consumer surplus. The implemation of souch a strategy on need a way to segment the market. The seller must be able to divide the market into groups based on some characteristics, in this assignment we have split them into summer and winter. This segmentation correlate with differences in price elasticity of demand. This difference allows the monopolist to adjust prices according to each segments sensitivity to changes. As well the seller must be able to prevent or limit the resale of goods from one segment to another. If consumers can easily resell it among themselves, they would buy the good at the lower price available in one segment and sell it in another, undermining the price discrimination strategy.

**Welfare Effects of Price Discrimination** \
Third-degree price discrimination significantly impacts the distribution of consumer and producer surplus across different market segments. By charging different prices to different segments based on their willingness to pay, a monopolist can capture a larger portion of the consumer surplus, converting it into producer surplus. This means that in segments with less price sensitivity (inelastic demand), consumers may end up paying more for the same good, reducing their consumer surplus. Conversely, in more price-sensitive segments (elastic demand), consumers might benefit from lower prices than what a uniform pricing strategy would offer. 

In segments where the price is reduced to target more price-sensitive consumers, consumer surplus can actually increase because the lower price makes the product accessible to consumers who would not purchase it at a higher uniform price. Overall the monopolist's ability to better extract consumer surplus by segmenting the market and pricing accordingly typically results in an increase in producer surplus. The firm captures additional revenue that under a single-price system would remain with consumers, or be lost due to lower sales volumes. This leads to a overall increase to total welfare as things are more efficient allocated. More consumers are able to purchase the product leads to reduction in deadweight loss compared to a single-price monopoly scenario, and this leads to a lose in deadweight loss. So third-degree price discrimination can, under certain conditions, mitigate this deadweight loss.

**Demand Changes** \
Demand can shift due to various factors: changes in consumer preferences, income levels, the prices of related goods, or seasonal changes. A demand increase (shift to the right) in one segment may allow a monopolist to raise prices without losing much in terms of volume, especially if the demand becomes less elastic. Or a shift to the left might requite lowering the price to maintain volumes, again especially if demand becomes more elastic. An increase in demand might also make the monopolist increase production to capture the additional consumer surplus assuming the marginal cost remain constant. A decrease would again have the opposite effect, reducing output.

To analyze the pricing strategy for the hotel and determine the optimal market adaptation for the hotel, we apply microeconomic theory, specifically the monopolist's approach to profit maximization through price discrimination.

For the winter seasn, wwill use the a linear demand functi we were given:\
$P = 7500 - 12.5Q$

This inverse demand function implies that as the price decreases by $12.5 NOK$, one additional room is demanded. Using this function, the total revenue ($TR$) for the winter season is modeled as:\
$TR = PxQ = (7500-12.5Q)Q$

To determin the quantity of rooms that maximizes profit we calculate the marginal revenue ($MR$) by taking the derivative of $TR$ relative to $Q$:\
$MR = \frac{d(TR)}{dQ} = 7500-25Q$

The profit-maximizing quantity is found where $MR$ equals marginal cost ($MC$), and as mention $MC$ is equal to $800 NOK$:\
$7500-25Q = 800 => Q = 268$

With the profit-maximizing quantity we can get the optimal price which is $4150 NOK$. 

With this we have optimal Q and P for winter:\
$Optimal P = 4150 NOK$\
$Optimal Q = 268$

And we can do the same calculation for summer just with the demand function given there was:\
$P = 2000 - 2.5Q$

And here we get:\
$Optimal P = 1400 NOK$\
$Optimal Q = 240$


With this we can look at a figure showing the optimal P and Q with demand curve and MR and MC as well.
